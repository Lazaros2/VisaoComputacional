```python
import cv2
import numpy as np

def remove_fundo_mediana(video, num_frames=50, output_path='video_sem_fundo.mp4'):
    cap = cv2.VideoCapture(video)
    
    # ler os frames para estimar o fundo
    frames = []
    for i in range(num_frames):
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)
        
    fundo = np.median(frames, axis=0).astype(np.uint8)

    # Configurar o VideoWriter para salvar o vídeo resultante
    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    codec = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, codec, fps, (width, height))

    while True:
        ret, frame = cap.read()

        if not ret:
            break

        # Calcular a diferença absoluta entre o quadro atual e o fundo fixo estimado
        diff = cv2.absdiff(frame, fundo)

        # Definir um limiar para identificar as partes alteradas do quadro
        threshold = 50
        mask = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
        _, mask = cv2.threshold(mask, threshold, 255, cv2.THRESH_BINARY)

        # Aplicar a máscara ao quadro para remover o fundo
        result = cv2.bitwise_and(frame, frame, mask=mask)

        # Gravar o quadro resultante no vídeo
        out.write(result)


    cap.release()
    out.release()
    cv2.destroyAllWindows()



```
